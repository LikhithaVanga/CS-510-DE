```python
import pandas as pd
import pandera as pa
import numpy as np
```

```python
df = pd.read_csv("Oregon Hwy 26 Crash Data for 2019 - Crashes on Hwy 26 during 2019.csv")
df.head(10)
```

```python
crashdf = df[(df['Record Type'] == 1)]
crashdf = crashdf.dropna(axis=1, how = 'all')
if crashdf.shape[0] == crashdf['Crash ID'].unique().shape[0]:
    print("all unique crashids-Validated")
print(crashdf.shape)
crashdf.drop_duplicates(keep=False,inplace=True)    
print(crashdf.shape)
# print(crashdf['Crash Year'])
```

```python
from collections import Counter
dayArray = []
montharray = []
weekarray = []
for row in range(len(crashdf)):
    year = int(crashdf['Crash Year'].iloc[row])
    month = int(crashdf['Crash Month'].iloc[row])
    day = int(crashdf['Crash Day'].iloc[row])
    weekday = int(crashdf['Week Day Code'].iloc[row])
    weekarray.append(weekday)
    montharray.append(month)
    dayArray.append(day)
    
dayCounter = Counter(dayArray)
monthcounter = Counter(montharray)
weekdaycounter = Counter(weekarray)
print("Months most crashes occured", monthcounter.most_common(5)[0][0])
print("Days many crashes occured", dayCounter.most_common(5))
print("Week day most crashes occured", weekdaycounter.most_common(2))
        
    
```

```python
crashTypeArray = []
for row in range(len(crashdf)):
    crashType = int(crashdf['Crash Type'].iloc[row])
    crashTypeArray.append(crashType)
crashtypeCounter = Counter(crashTypeArray)
print("Most common carsh types", crashtypeCounter.most_common(2))
    
```
